This tutorial demonstrates how to use the Aneto_Utils to generate input files from vasp output files.

```python
!pip install ase pytest numpy
import aneto_utils as au
import numpy as np
import os
```

We will be using files which can be found ./test_files for this demo.

```python
ls test_files/
```

First, we load the reference data from the test_files and the elastic modulus data

```python
ref_data = au.load_ref_data('./test_files')
e_mod = au.read_cij_json('cij.json')
#print(ref_data.keys())
```

Next we select a directory with defect data in "OUTCAR" and generate aneto input files with this data and already loaded reference and elastic modulus data. This file will be generated in the default input file directory with the label "_test". The new file will be found at './aneto_input_files/input_elast_test'. If you have multiple defects, here you can point to different defect directories and label each file accordingly.

```python
au.setup_aneto('./test_files/',ref_data, e_mod)
```

We can now check the file location and print its contents.

```python
os.system('ls ./aneto_input_files')
os.system('cat aneto_input_files/input_elast__test')
    
```
